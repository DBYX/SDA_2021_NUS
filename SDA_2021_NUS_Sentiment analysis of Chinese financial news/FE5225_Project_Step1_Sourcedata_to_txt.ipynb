```python
import pandas as pd
import numpy as np
```

```python
# load the data and transfer the data to txt file for furthur use
# there may be 5 excel under each year
idx = ["","_1","_2","_3","_4"]
# create two empty series for storage
pos_data_all = pd.Series()
neg_data_all = pd.Series()

for i in range(1,21):
    for j in range(5):
        #print(i,j)
        print("progress:",(i-1)*5+(j+1),"%")
        try:
            year = str(i).zfill(2)
            data_path = "./source_data/20"+year+"/网络财经新闻基本信息-20"+year+idx[j]+".xlsx"
            data = pd.read_excel(data_path,header=0,error_bad_lines=False,encoding = "utf-8")
            pos_data = data["Newstitle"][data["Newsemot"]=="1"]
            #print(pos_data)
            neg_data = data["Newstitle"][data["Newsemot"]=="-1"]
            #print(neg_data)
            pos_data_all = pos_data_all.append(pos_data,ignore_index = True)
            #print(pos_data_all)
            neg_data_all = neg_data_all.append(neg_data,ignore_index = True)
            #print(neg_data_all)
        except FileNotFoundError:
            print("skip year:",i,"index:",j) # skip the data_path which is not correct
            continue
# the result skip most of the all,because I only put one data file in the source directory for testing
```

```python
#print several data to see whether the format is correct
print(pos_data_all[0:5])
print(neg_data_all[0:5])
```

```python
#save the data to .txt file
pos_data_all.to_csv('./pos_data0331.txt',index=False,header = None,encoding='UTF-8')
neg_data_all.to_csv('./neg_data0331.txt',index=False,header = None,encoding='UTF-8')
```
