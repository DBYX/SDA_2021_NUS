```python
# import all the related library
import numpy as np
import snownlp
from snownlp import SnowNLP
from snownlp import sentiment
import os
import codecs
import pandas as pd
import re
import numpy as np
from pandas import DataFrame
```

```python
# save the list as an excel file
def writeexcel(list,address):
    df=DataFrame(list)
    df.to_excel(address)
    print('done')

# calculate the sentiment 
def gradeemotion(text):
    score=[]
    num=0
    s = SnowNLP(text)
    sentence=s.sentences
    for x in sentence:
        emotion=SnowNLP(x).sentiments
        score.append(emotion)
        if emotion<0.5:
            num=num+1
    ave=np.mean(score)
    return ave
```

```python
testmews='./sina_news.csv'
df = pd.read_csv(testmews,index_col = 0,encoding = "utf-8",engine="python")

news=[]  #time+new
test=[]
for i in range(len(df)):
    a=[]
    date1=df.iloc[i,0].split(" ")
    date=date1[0]
    try:
        new=df.iloc[i,1]   
        a.append(date)
        a.append(new)
        test.append(new)
        news.append(a)
    except:
        continue
```

```python
num = len(news)
for k in range(len(news)):
    try:       
        score=gradeemotion(news[k][1])
        news[k].append(score)
        print("Progressing:{:.2f}%".format(k/len(news)*100))
    except:
        continue

writeexcel(news,'./news_score_model4.xlsx')
```

```python
s=pd.read_excel('./news_score_model4.xlsx',index_col=0)
sum=0
date=[]
count=0
average_score=[]
for i in range(len(s)):
    if s.iloc[i,0] not in date:
        date.append(s.iloc[i,0])
 
```

```python
for k in range(len(date)):
    perday=[]
    sum=0
    count=0
    perday.append(date[k])
    print("Progressing:{:.2f}%".format(k/len(date)*100))
    for u in range(len(s)):
        if s.iloc[u,0]==date[k]:
            sum=sum+s.iloc[u,2]
            count=count+1
    ave=sum/count
    perday.append(ave)
    average_score.append(perday)

writeexcel(average_score,'./average_score_perday.xlsx')   
```

```python
std_score=[]
for k in range(len(date)):
    perday=[]
    sum=[]
    count=0
    perday.append(date[k])
    print("Progressing:{:.2f}%".format(k/len(date)*100))
    for u in range(len(s)):
        if s.iloc[u,0]==date[k]:
            sum.append(s.iloc[u,2])
            count=count+1
    ave=np.std(sum,ddof=1)
    perday.append(ave)
    std_score.append(perday)
     
writeexcel(std_score,'./std_score_perday.xlsx')
```
